Overall analysis:
In order to gain an overall analysis on the current world's agricultural production, we can use the gapminder plot for the most recent year. This plot is based on the infamous gapminder tools. 

```python
# find the most recent year
most_recent_year = showcase_obj.df['Year'].max()

# gapminder
print("Gapmionder plot for the most recent year: ")
showcase_obj.gapminder(int(most_recent_year))
```


From the gapminder plot we can see that `output_quantity` increases linearly with `fertilizer_quantity`. We can also see that as (both) increase, the `animal_output_quantity` also increases, this is indicated by the area of each dot.

This means that agricultural production in the world is heavily dependent on the quantity of fertilizer used. As fertilizer quantity increases, both output quantity and animal output quantity also tend to increase, implying a positive relationship between fertilizer use and agricultural productivity. This suggests that there may be room for increased use of fertilizers in agriculture to boost production.

Furthermore, the fact that the size of the dots increases with both x and y values indicates that there are regions with high levels of agricultural productivity and animal output, potentially indicating areas with more advanced agricultural practices and infrastructure. However, it is also important to consider potential environmental and social impacts of increased fertilizer use and explore more sustainable approaches to agricultural productivity.


First we consider the normalized versions of the graphs, in order to understand the differences in the output of each distinct `_output_` column. For Germany, the fish output is similar to that of the US, with it being relatively insignificant. Animal output exceeds crop output, by up to around 30% at some stages, so much more than the US (in terms of percentage of total output for each respective country). Lastly, for Japan, the fish output is much larger than the formerly mentioned countries, up to around 10% of total output for Japan. In a converse way to the US, initially Japan had crop output exceeding animal output by up to 60% but this changed in around 1990, where animal output overtook crop output. 

When comparisons are made here, they are made in references to percentage output per that distinct country, not in absolute terms of comparison of output quantity. 


Here, in order to compare the countries to each other, it is more intuitive to compare the normalized versions of the graphs, then the actual values of the output. Next, we consider the choropleth for the most recent year (which currently is 2019), and 2000. 

```python
showcase_obj.choropleth(int(most_recent_year))
```

```python
showcase_obj.choropleth(2000)
```

Considering the above Choropleth total factor productivity (TFP) maps for the most recent year, which in this case is 2019, and for 2000. We can see some stark differences. 

The first thing to notice is that the scales are different. In 2000, the TFP in the world covered a wider range of values, and ranges from 50 to 250, while in 2019, the TFP ranges between a smaller range of around 80 to 160. This shows us that the general trends in the world are changing with time, but maybe not in the way we would expect. However, upon closer inspection of the graph, there are no major countries in 2000 which exceed a TFP of around 150. Then considering 2019, we see that there are not many (easily visible) countries with a TFP below 120. So from this we can make a general assumption that the TFP increases with time. 
However, from the maps it is not clear how specific countries evolve, and that some countries are less visible on the map. For this reason, we can analyses the countries TFP, for specific countries, and also predict on how they would change in the future. We do this using the predictor method. 

```python
countries_predictor = ["United States of America", "Germany", "Japan"]
showcase_obj.predictor(countries_predictor)
```

Lastly, we consider the TFP for the three countries, United States, Germany and Japan. Here it seems that these three countries follow the same general trend for the years up to 2019. Here, we can see clearly that the TFP increases with time, in a general sense, with small fluctuations between the years. The forecast for the coming years yields some interesting conclusions though. Japan had had the lowest TFP of the three countries until 2019, but is predicted to overtake the other countries in the coming years. While conversely for the USA, which had the highest TFP, is predicted to have the lowest TFP of the three in the coming years. This is an interesting observation, but may be related to the change in growth rates in the last decades, which has been increasing for Japan, but decreasing for the USA.
